# Loading Packages and Data

```python
import numpy as np
import pandas as pd
import sklearn.metrics as sm
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report
```

## Data Preprocessing

```python
target=data['Species']
df=data.copy()
df=df.drop('Species', axis=1)
df.shape
```

```python
x=data.iloc[:, [0,1,2,3]].values
LaEn=LabelEncoder()
data['Species']=LaEn.fit_transform(data['Species'])
y=data['Species'].values
data.shape
```

## Training the Model

### Spliting the data into Training and Testing set

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("Trainging set:",x_train.shape)
print("Testing set:",x_test.shape)
```

### Defining decision tree Algorithm

```python
d_tree=DecisionTreeClassifier()
d_tree.fit(x_train,y_train)
print("Decision Tree Classifier created!")
```

## Visualisation of trained model

```python
plt.figure(figsize=(30,20))
tree=plot_tree(d_tree,feature_names=df.columns,precision=2,rounded=True,filled=True,class_names=target.values)
```

## Classification Report

```python
y_pred=d_tree.predict(x_test)
print("Classification Report:\n",classification_report(y_test,y_pred))
```

## Accuracy

```python
print("Accuracy :", sm.accuracy_score(y_test,y_pred))
```

## Confusion Matrix

```python
confusion_matrix(y_test,y_pred)
```
